```python
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
RecallRate = pd.read_csv(r"F:\\24561\\Documents\\欧贝杯智能创新比赛\\数据\\Weight&Recalltrain.csv")
RecallRate = pd.DataFrame(data=RecallRate)
```

```python
def plot_concrete_entre_dischange(RecallRate,entre):
    chooseweight = RecallRate[RecallRate['Entre']==entre]
    weightDis = np.asarray(chooseweight.iloc[:,1])
    recallObject = np.asarray(chooseweight.iloc[:,3])
    recallObject = recallObject.astype('float64')

    plt.plot(weightDis,recallObject,color="green",label="Recallrate")
    plt.xticks(weightDis)
    plt.xticks(rotation = 45)
    plt.legend()
    plt.xlabel("距离的权重")
    plt.ylabel("训练集上召回率")
    plt.title("训练集上固定企业资质权重为{}时召回率随距离权重的变化的学习曲线".format(entre))
    plt.show()
```

```python
entre = 0.3
plot_concrete_entre_dischange(RecallRate,entre)
```
